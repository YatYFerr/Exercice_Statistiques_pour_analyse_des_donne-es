```python
import pandas as pd
```

```python
# Assiette salariale de 13 employés
entreprise1 = {
    'assistant1': 1000,
    'assistant2': 1080,
    'analyste1': 1500,
    'analyste2': 1520,
    'analyste3': 1600,
    'analyste4': 1670,
    'manager1': 2640,
    'manager2': 2730,
    'manager3': 2710,
    'directeur1': 5000,
    'directeur2': 5000,
    'CEO': 57000
}

entreprise2 = {
    'assistant1': 3000,
    'assistant2': 3000,
    'analyste1': 4200,
    'analyste2': 4200,
    'analyste3': 4200,
    'analyste4': 4200,
    'manager1': 6000,
    'manager2': 6000,
    'manager3': 6000,
    'directeur1': 8000,
    'directeur2': 8000,
    'CEO': 15000
}
```

```python
entreprise1 = pd.DataFrame(entreprise1.items(), columns=['Poste', 'Salaire'])
entreprise2 = pd.DataFrame(entreprise2.items(), columns=['Poste', 'Salaire'])
```

## Moyenne

```python
moyenne_salaire1 = entreprise1.Salaire.mean()
print('Entreprise 1 - Salaire: {:.2f}'.format(moyenne_salaire1))
moyenne_salaire2 = entreprise2.Salaire.mean()
print('Entreprise 2 - Salaire: {:.2f}'.format(moyenne_salaire2))
```

## Médiane

```python
mediane_salaire1 = entreprise1.Salaire.median()
print('Entreprise 1 - Salaire: {:.2f}'.format(mediane_salaire1))
mediane_salaire2 = entreprise2.Salaire.median()
print('Entreprise 2 - Salaire: {:.2f}'.format(mediane_salaire2))
```

## Mode

```python
mode_salaire1 = entreprise1.Salaire.mode()[0]
print('Entreprise 1 - Salaire: {:.2f}'.format(mode_salaire1))
mode_salaire2 = entreprise2.Salaire.mode()[0]
print('Entreprise 2 - Salaire: {:.2f}'.format(mode_salaire2))
```

# Visualization de la moyenne, médiane et mode

```python
import matplotlib.pyplot as plt
```

```python
valeurs = {
    'list1': [1,5,7,9,100] # valeurs: 5, 7 -> 9 -> 1 -> 100
}

valeurs = pd.DataFrame(valeurs)
```

```python
moyenne = valeurs.list1.mean()

plt.hist(valeurs.list1);
plt.axvline(x=moyenne, color='r', linestyle='-')
plt.show()

print('moyenne: '+ str(moyenne))
```

```python
valeurs = {
    'list2': [1,5,7,7,9,13,100] # valeurs: 5, 7 -> 7 -> 9 -> 1 -> 13 -> 100
}

valeurs = pd.DataFrame(valeurs)
```

```python
moyenne = valeurs.list2.mean()
mediane = valeurs.list2.median()
mode = valeurs.list2.mode()
ecart_type = valeurs.list2.std()

plt.hist(valeurs.list2);
plt.axvline(x=moyenne, color='r', linestyle='-')
plt.axvline(x=mediane, color='b', linestyle='-')
plt.axvline(x=mode[0], color='g', linestyle='-')
plt.show()

print('moyenne: '+ str(moyenne))
print('mediane: '+ str(mediane))
print('mode: '+ str(mode))
print('écart-type: '+ str(ecart_type))
```

```python
x = range(0,len(valeurs.list2))

plt.scatter(x,valeurs.list2)
plt.axhline(y=moyenne, color='r', linestyle='-')
plt.show()
```
